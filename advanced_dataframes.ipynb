```python
# 5. Load the mpg dataset from PyDataset.

from pydataset import data
mpg = data('mpg')
```

```python
# 6. Output and read the documentation for the mpg dataset.

mpg
```

```python
# 7. How many rows and columns are in the dataset?

mpg.shape
```

```python
# 8. Check out your column names and perform any cleanup you may want on them.

mpg.rename(columns={'cty': 'city'}, inplace=True)
mpg.rename(columns={'hwy': 'highway'}, inplace=True)
mpg.rename(columns={'drv': 'drive'}, inplace=True)
mpg.rename(columns={'displ': 'eng_displ_litres'}, inplace=True)
mpg
```

```python
# 9. Display the summary statistics for the dataset.

mpg.describe()
```

```python
# 10. How many different manufacturers are there?

num_manufacturer = mpg.manufacturer.unique()
len(num_manufacturer)
```

```python
# 11. How many different models are there?

num_model = mpg.model.unique()
len(num_model)
```

```python
# 12. Create a column named mileage_difference like you did in the 
# DataFrames exercises; this column should contain the difference 
# between highway and city mileage for each car.

mpg['mileage_difference'] = (mpg.highway - mpg.city)
mpg
```

```python
# 13. Create a column named average_mileage like you did in the DataFrames 
# exercises; this is the mean of the city and highway mileage.

mpg['average_mileage'] = ((mpg.city + mpg.highway)/2)
mpg
```

```python
# 14. Create a new column on the mpg dataset named is_automatic that 
# holds boolean values denoting whether the car has an automatic transmission.

mpg['is_automatic'] = mpg.trans.str.startswith('a')
mpg
```

```python
# 15. Using the mpg dataset, find out which which manufacturer has the 
# best miles per gallon on average?

manu_mpg_avg = mpg.groupby(['manufacturer']).average_mileage.agg(['mean'])
manu_mpg_avg['mean'].nlargest(n=1)
```

```python
# 16. Do automatic or manual cars have better miles per gallon?

mpg.groupby(['is_automatic']).average_mileage.agg(['mean'])
# manual cars have better miles per gallon
```

```python
# Exercises III
# 1. Use your get_db_url function to help you explore the data from the chipotle database.

url = f'mysql+pymysql://{user}:{password}@{host}/chipotle'
orders = pd.read_sql('SELECT * FROM orders;', url)
```

```python
# 2. What is the total price for each order?

item_price = orders['item_price'].str.replace('$', '')
orders['price_float'] = item_price.astype('float')
orders
orders.groupby(['order_id']).price_float.agg(['sum'])
```

```python
# 3. What are the most popular 3 items?

orders['item_name'].value_counts().nlargest(n=3)
```

```python
# 4. Which item has produced the most revenue?

orders.groupby(['item_name']).price_float.agg(['sum'])['sum'].nlargest(n=1)
```

```python
# 5. Join the employees and titles DataFrames together.

emp_title = emp_table.merge(titles_table, left_on='emp_no', right_on='emp_no', how='inner', indicator=True)
```

```python
# 6. For each title, find the hire date of the employee that was hired most recently with that title.

emp_title.groupby(['title']).hire_date.agg(['max'])
```

```python
# 7. Write the code necessary to create a cross tabulation of the 
# number of titles by department. (Hint: this will involve a combination 
# of SQL code to pull the necessary data and python/pandas code to 
# perform the manipulations.)

query_2 = 'SELECT title, dept_name FROM dept_emp JOIN titles USING (emp_no)\
JOIN departments USING (dept_no);'
title_dept = pd.read_sql(query_2, url_emp)

pd.crosstab(title_dept.title, title_dept.dept_name)
```
