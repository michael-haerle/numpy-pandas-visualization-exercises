```python
# 1. Use matplotlib to plot the following equation:
# y= x^2−x+2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

import matplotlib.pyplot as plt
x = list(range(-150, 150))
y = [(n**2)-(n + 2) for n in x]
plt.plot(x, y)
plt.title('$y = x^2 - x+2$')
plt.annotate('Origin', xy=(0,0), xytext=(-25,2500), arrowprops={'facecolor':'red'})
plt.show
```

```python
# using np to solve
import numpy as np
x = list(range(-150, 150))
x = np.array(x)
y = [(n**2)-(n + 2) for n in x]
plt.plot(x, y)
plt.title('$y = x^2 - x+2$')
plt.annotate('Origin', xy=(0,0), xytext=(-25,2500), arrowprops={'facecolor':'red'})
plt.show
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y=√x
import math
x = np.array(range(150))
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
plt.show
# You can use functions from the math module to help implement some of the equations above.
```

```python
# y=x^3
x = range(150)
x = np.array(x)
y = [(n**3) for n in x]
plt.plot(x, y)
plt.show
```

```python
# y=2^x
x = range(150)
x = np.array(x)
y = [(2**n) for n in x]
plt.plot(x, y)
plt.show
```

```python
# y=1/(x+1)
x = range(150)
x = np.array(x)
y = [1/(n + 1) for n in x]
plt.plot(x, y)
plt.show
```

```python
# 3. Combine the figures you created in the last step into 
# one large figure with 4 subplots.

import matplotlib.figure as fig
plt.fig, axs = plt.subplots(2, 2)

x = np.array(range(150))
y = [math.sqrt(n) for n in x]
axs[0, 0].plot(x, y)

a = [(n**3) for n in x]
axs[1, 0].plot(x, a)

b = [(2**n) for n in x]
axs[0, 1].plot(x, b)

c = [1/(n + 1) for n in x]
axs[1, 1].plot(x, c)

plt.show
```

```python
# 4. Combine the figures you created in the last step into one figure 
# where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.
fig = plt.figure()
fig.suptitle("Four Graphs in One Fig")

x = np.array(range(150))
y = [math.sqrt(n) for n in x]
plt.subplot(221)
plt.plot(x, y, 'tab:green', label = '$y=√x$')

a = [(n**3) for n in x]
plt.subplot(222)
plt.plot(x, a, 'tab:red', label = '$y=x^3$')

b = [(2**n) for n in x]
plt.subplot(223)
plt.plot(x, b, 'tab:orange', label = '$y=2^x$')

c = [1/(n + 1) for n in x]
plt.subplot(224)
plt.plot(x, c, label = '$y=1/(x+1)$')

fig.legend()
plt.show
```
